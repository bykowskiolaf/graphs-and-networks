```python
class AdjacencyMatrix:
    def __init__(self, size=0):
        # Tworzy początkową macierz size x size wypełnioną zerami
        self.matrix = [[0] * size for _ in range(size)]

    def add_vertex(self):
        # Jeśli macierz jest pusta, tworzymy 1x1
        if not self.matrix:
            self.matrix.append([0])
            print("Dodano pierwszy wierzchołek.")
            return
            
        # Dodajemy zera na końcu każdego istniejącego wiersza (nowa kolumna)
        for row in self.matrix:
            row.append(0)
            
        # Dodajemy nowy wiersz na samym dole wypełniony zerami (o długości takiej, jak pierwszy wiersz)
        self.matrix.append([0] * len(self.matrix[0]))
        print(f"Dodano wierzchołek. Aktualny rozmiar: {len(self.matrix)}")

    def remove_vertex(self, v):
        if 0 <= v < len(self.matrix):
            # Usuwamy cały wiersz
            self.matrix.pop(v)
            # Usuwamy v-tą kolumnę z każdego pozostałego wiersza
            for row in self.matrix:
                row.pop(v)
            print(f"Usunięto wierzchołek {v}.")

    def add_edge(self, u, v):
        if 0 <= u < len(self.matrix) and 0 <= v < len(self.matrix):
            self.matrix[u][v] = 1
            self.matrix[v][u] = 1 # Zakładamy graf nieskierowany

    def remove_edge(self, u, v):
        if 0 <= u < len(self.matrix) and 0 <= v < len(self.matrix):
            self.matrix[u][v] = 0
            self.matrix[v][u] = 0

    def display(self):
        for row in self.matrix:
            print(row)
        print("-" * 15)


class AdjacencyList:
    def __init__(self):
        self.adj_list = {}

    def add_vertex(self, v):
        if v not in self.adj_list:
            self.adj_list[v] = []

    def remove_vertex(self, v):
        if v in self.adj_list:
            # Najpierw usuwamy ten wierzchołek z list sąsiadów
            for neighbor in self.adj_list[v]:
                self.adj_list[neighbor].remove(v)
            # Potem usuwamy sam wierzchołek z głównego słownika
            del self.adj_list[v]

    def add_edge(self, u, v):
        self.add_vertex(u)
        self.add_vertex(v)
        if v not in self.adj_list[u]:
            self.adj_list[u].append(v)
        if u not in self.adj_list[v]:
            self.adj_list[v].append(u)

    def remove_edge(self, u, v):
        if u in self.adj_list and v in self.adj_list[u]:
            self.adj_list[u].remove(v)
        if v in self.adj_list and u in self.adj_list[v]:
            self.adj_list[v].remove(u)

    def display(self):
        for k, v in self.adj_list.items():
            print(f"{k}: {v}")
        print("-" * 15)


# --- TRANSFORMACJE ---

def matrix_to_list(matrix_obj):
    list_obj = AdjacencyList()
    mat = matrix_obj.matrix
    for i in range(len(mat)):
        list_obj.add_vertex(i)
        for j in range(len(mat[i])):
            if mat[i][j] == 1:
                list_obj.add_edge(i, j)
    return list_obj

def list_to_matrix(list_obj):
    # Zakładamy, że wierzchołki to kolejne liczby całkowite od 0 do n-1
    nodes = list(list_obj.adj_list.keys())
    if not nodes:
        return AdjacencyMatrix(0)
    
    max_node = max(nodes)
    matrix_obj = AdjacencyMatrix(max_node + 1)
    
    for u in list_obj.adj_list:
        for v in list_obj.adj_list[u]:
            matrix_obj.add_edge(u, v)
    return matrix_obj

# --- SZYBKI TEST ---
print("Test Macierzy:")
m = AdjacencyMatrix(3) # Macierz 3x3
m.add_edge(0, 1)
m.add_edge(1, 2)
m.display()

m.add_vertex() # Dodaje 4. wierzchołek (indeks 3)
m.add_edge(0, 3)
m.display()

print("Konwersja na Listę:")
l = matrix_to_list(m)
l.display()
```

```python
import time
import sys
import random

def memory_and_time_test(vertices, edges_prob):
    print(f"--- TEST DLA {vertices} WIERZCHOŁKÓW ({edges_prob * 100:.0f}%) ---")
    
    # 1. Test Macierzy
    start_time = time.time()
    matrix_graph = AdjacencyMatrix(vertices)
    edges_count = 0
    
    for i in range(vertices):
        for j in range(i + 1, vertices):
            if random.random() < edges_prob:  # Prawdopodobieństwo utworzenia krawędzi
                matrix_graph.add_edge(i, j)
                edges_count += 1
                
    matrix_time = time.time() - start_time
    # sys.getsizeof() sprawdza rozmiar samej listy zewnętrznej, musimy zsumować wiersze
    matrix_memory = sys.getsizeof(matrix_graph.matrix) + sum(sys.getsizeof(row) for row in matrix_graph.matrix)
    
    print(f"Macierz: Czas generowania = {matrix_time:.4f} s")
    print(f"Macierz: Zużycie pamięci (szacunkowe) = {matrix_memory / 1024:.2f} KB")

    # 2. Test Listy
    start_time = time.time()
    list_graph = AdjacencyList()
    
    # Kopiujemy te same krawędzie z macierzy do listy, żeby test był sprawiedliwy
    for i in range(vertices):
        list_graph.add_vertex(i)
        for j in range(i + 1, vertices):
            if matrix_graph.matrix[i][j] == 1:
                list_graph.add_edge(i, j)
                
    list_time = time.time() - start_time
    # Szacujemy rozmiar słownika i wszystkich list wewnątrz
    list_memory = sys.getsizeof(list_graph.adj_list) + sum(sys.getsizeof(v) for v in list_graph.adj_list.values())
    
    print(f"Lista:   Czas generowania = {list_time:.4f} s")
    print(f"Lista:   Zużycie pamięci (szacunkowe) = {list_memory / 1024:.2f} KB")
    print(f"Wygenerowano krawędzi: {edges_count} \n")


# Uruchamiamy test dla 1000 wierzchołków i 10% szansy na krawędź
memory_and_time_test(1000, 0.1)
memory_and_time_test(1000, 0.1)
memory_and_time_test(5000, 0.1)
```

```python
class GraphList:
    def __init__(self):
        # Reprezentacja wewnętrzna: Słownik, gdzie kluczem jest ID wierzchołka, a wartością lista sąsiadów
        self.adj_list = {}

    def add_vertex(self, v):
        if v not in self.adj_list:
            self.adj_list[v] = []

    def remove_vertex(self, v):
        if v in self.adj_list:
            # Najpierw usuwamy ten wierzchołek z list sąsiadów
            for neighbor in self.adj_list[v]:
                self.adj_list[neighbor].remove(v)
            # Następnie usuwamy sam wierzchołek
            del self.adj_list[v]

    def add_edge(self, u, v):
        # Upewniamy się, że wierzchołki istnieją
        self.add_vertex(u)
        self.add_vertex(v)
        # Dodajemy krawędź nieskierowaną (w obie strony)
        if v not in self.adj_list[u]:
            self.adj_list[u].append(v)
        if u not in self.adj_list[v]:
            self.adj_list[v].append(u)

    def remove_edge(self, u, v):
        if u in self.adj_list and v in self.adj_list[u]:
            self.adj_list[u].remove(v)
        if v in self.adj_list and u in self.adj_list[v]:
            self.adj_list[v].remove(u)

class Rectangle:
    def __init__(self, id, x1, y1, x2, y2):
        self.id = id
        self.x1 = float(x1)
        self.y1 = float(y1)
        self.x2 = float(x2)
        self.y2 = float(y2)

    def intersects(self, other):
        # Dwa prostokąty NIE przecinają się, jeśli jeden jest całkowicie z boku, nad lub pod drugim
        return not (self.x2 < other.x1 or
                    self.x1 > other.x2 or
                    self.y2 < other.y1 or
                    self.y1 > other.y2)

def load_rectangles_and_build_graph(filename):
    rectangles = []
    graph = GraphList()

    # Wczytywanie prostokątów z pliku
    with open(filename, 'r') as f:
        for line in f:
            parts = line.split()
            if len(parts) == 5:
                rect_id = parts[0]
                x1, y1, x2, y2 = parts[1], parts[2], parts[3], parts[4]
                rectangles.append(Rectangle(rect_id, x1, y1, x2, y2))
                graph.add_vertex(rect_id) # Dodajemy od razu wierzchołek do grafu

    # Sprawdzanie przecięć i dodawanie krawędzi
    for i in range(len(rectangles)):
        for j in range(i + 1, len(rectangles)):
            if rectangles[i].intersects(rectangles[j]):
                graph.add_edge(rectangles[i].id, rectangles[j].id)

    return graph

# --- PRZYKŁAD UŻYCIA ---
g = load_rectangles_and_build_graph('prostokaty.txt')
print("Listy sąsiedztwa:", g.adj_list)
```

```python
# macierz sąsiedztwa grafu G (nieskierowany)
G =       [[0, 0, 1, 1],
          [0, 0, 1, 1],
          [1, 1, 0, 1],
          [1, 1, 1, 0]]

# liczba krawędzi
def nEdges(G):
  m = 0
  n = len(G[0]) # liczba wierzchołków
  for row in range(n):
    for col in range(n):
      if(G[row][col] == 1):
        m += 1
  return (int)(m/2) # graf nieskierowany wymaga podzielenia liczby krawędzi przez 2, ponieważ każdą krawędź policzyliśmy 2 razy

# graf krawędziowy grafu G
def lineGraph(G):
    if(G == []): # graf pusty - kończymy
        return
    n = len(G[0]) # liczba wierzchołków
    m = nEdges(G) # liczba krawędzi
    H = [[0 for row in range(m)] for col in range(m)]

    pairs = []
    # przeglądam listę sąsiedztwa grafu G i tworzę listę krawędzi = par wierzchołków
    for row in range(n):
        # przeglądam tylko górną połowę (powyżej ukośnej linii wiersz = kolumna) - zakładam graf nieskierowany
        for col in range(row, n):
            if(G[row][col] == 1):
                pairs.append([row, col])

    # przeglądam listę par...
    for i in range(len(pairs)):
        for v in pairs[i]:
            for j in range(i + 1, len(pairs)):
                # ...i jeżeli znajdę wspólny wierzchołek, zapisuję 1 w odpowiednim indeksie macierzy sąsiedztwa grafu krawędziowego H
                if v in pairs[j]:
                    # symetrycznie, bo pracujemy na grafie nieskierowanym
                    H[i][j] = 1
                    H[j][i] = 1
    return H

# dopełnienie grafu G
def complement(G):
    if(G == []): # graf pusty - kończymy
        return
    n = len(G[0]) # liczba wierzchołków
    H = [[0 for row in range(n)] for col in range(n)] # szykujemy (na razie wyzerowaną) macierz sąsiedztwa dopełnienia
    for row in range(n):
        for col in range(n):
            if(G[row][col] == 1): # jeśli w grafie była krawędź, to w dopełnieniu nie będzie...
                H[row][col] = 0
            else:                 # i vice versa
                H[row][col] = 1
    return H

print("Macierz sąsiedztwa grafu krawędziowego grafu G:", lineGraph(G))
print("Macierz sąsiedztwa dopełnienia grafu G:", complement(G))
```

```python
# macierz sąsiedztw
matrix = [[0, 1, 1, 1],
          [1, 0, 2, 2],
          [1, 2, 0, 0],
          [1, 2, 0, 0]]

# liczenie stopnia
def deg(matrix, node):
    n = len(matrix[0]) # liczba wierzchołków
    nodeIdx = node - 1  # zamiana nr wierzcholka na indeks w tablicy
    if nodeIdx < n: # sprawdzenie, czy node istnieje
        deg = 0
        for neighbor in range(n):
            deg += matrix[nodeIdx][neighbor]
        return deg
    else:
        return 0

# potencjał - liczba wierzchołków w sąsiedztwie wierzchołka, których stopień jest niemniejszy od ich numeru
def p(matrix, node):
    n = len(matrix[0]) # liczba wierzchołków
    nodeIdx = node - 1  # zamiana nr wierzcholka na indeks w tablicy
    k = 0
    if nodeIdx < n: # sprawdzenie, czy node istnieje
      for neighbor in range(nodeIdx, n): # sprawdzamy sąsiadów
        if(deg(matrix, neighbor + 1) >= neighbor + 1): # zamiana indeksu na nr wierzchołka
          k += 1
    return k

print(p(matrix, 1))
```
